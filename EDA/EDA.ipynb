## What is the transaction for each month / product

```python
df_transaction['InvoiceDate'] = pd.to_datetime(df_transaction['InvoiceDate'])
df_transaction.groupby(df_transaction['InvoiceDate'].dt.strftime('%B'))['NetAmount'].sum()
```

```python
grouped_products = pd.DataFrame(df_transaction.groupby('GroupNameLevel1').sum()['NetAmount'].sort_values(ascending=False))
grouped_products['percent'] = (df_transaction.groupby('GroupNameLevel1').sum()['NetAmount'] / df_transaction['NetAmount'].sum()) * 100
grouped_products
```

Top 5 products based on net amount:
1. M150
2. C-VITT
3. LIPO
4. Som in Sum
5. Chalarm

Clearly there's a outlier product, M-150 taking 46.2% of majority of profit
